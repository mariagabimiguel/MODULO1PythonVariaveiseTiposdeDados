####Para ajudar a quebrar o problema em etapas temos alguns dados a serem pedidos para que seja possível chegar ao resultado da proposta:

* Qual é a porcentagem da ineficácia da vacina? 
* Qual a porcentagem de meninas que são acometidas pela doença por não serem vacinadas?
* Qual é a porcentagem máxima estabelecida para a probabilidade de escolha ao acaso de uma menina vir a desenvolver a doença para a proposta de vacinação a ser realizada?

#### Com isso, temos dois cenários: 
(para aplicar o valor da probabilidade de uma menina escolhida vir a desenvolver a doença e obtermos o valor da cobertura de vacinação para o caso)

* Grupo doentes sem vacina
* Grupo doentes com vacina mas ineficaz

```python
ineficacia = float(input("Qual é a porcentagem da ineficácia da vacina? "))
```

```python
ineficacia = (ineficacia/100)
```

```python
print (ineficacia)
```

####A ideia de usar a porcentagem do público não vacinado aqui é obter o dado das meninas acometidas pela doença por não tomarem a vacina: será o grupo doentes não vacinadas = (1 - x )*doentes_naovacinadas, sendo (1 - x) o valor conjugado do público vacinado (x) 

```python
doentes_naovacinadas = float(input("Qual a porcentagem de meninas que são acometidas pela doença por não serem vacinadas? "))
```

```python
vacinadas = (100 - doentes_naovacinadas)
```

####A ideia de usar a porcentagem do público vacinado aqui é obter o dado das meninas acometidas pela doença por ineficácia da vacina: será o grupo doentes vacinadas = (vacinadas * ineficacia * x), sendo x o valor do público vacinado

```python
print (vacinadas)
```

```python
doentes_naovacinadas = (doentes_naovacinadas/100)
```

```python
print (doentes_naovacinadas)
```

```python
vacinadas = (vacinadas/100)
```

```python
print (vacinadas)
```

```python
meta_doentes = float(input ("Qual é a porcentagem máxima estabelecida para a probabilidade de escolha ao acaso de uma menina vir a desenvolver a doença para a proposta de vacinação a ser realizada? "))
```

```python
meta_doentes = (meta_doentes / 100)
```

```python
print (meta_doentes)
```

###EQUAÇÃO SIMPLES 

```python
#((1-x)*doentes_naovacinadas) + (vacinadas*ineficacia*x) = meta_doentes
```

### Escolhi usar a biblioteca sympy para aplicar os valores na equação:


```python
!pip install sympy
```

```python
import sympy
```

```python
x = sympy.Symbol('x')
```

```python
equation = sympy.Eq(((1-x)*doentes_naovacinadas) + (vacinadas*ineficacia*x), meta_doentes)
```

```python
equation
```

```python
proposta = sympy.solve(equation, x)
```

```python
print(proposta)
```
